```python
#import libraries and import the csv file
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

filename=pd.read_csv('Telco-Customer-Churn.xlsx - Telco-Customer-Churn.csv')
df=pd.DataFrame(filename)

import warnings
warnings.filterwarnings("ignore")
```

```python
df
```

# 1.Compare churn count with respect to gender?

```python
df['gender'].value_counts()
```

```python
plt.figure(figsize=(6,8))
ax=sns.countplot(df['Churn'],hue=df['gender'])
plt.title('Churn Count with respect to gender',fontsize=15)
plt.legend( bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.xlabel('Churn',fontsize=14)
plt.ylabel('Count',fontsize=14)

for p in ax.patches:
    ax.text(p.get_x()+.07,p.get_height()-200,str(p.get_height()),fontsize=18,color='white',fontweight='bold')
```

# 2.Find out how many female senior citizens there in the dataset

```python
data1=df[(df['SeniorCitizen']==1)]
data1['SeniorCitizen'].value_counts()
```

```python
data1=df[(df['SeniorCitizen']==1)]
ax=sns.countplot(data1['SeniorCitizen'],hue=df['gender'])
plt.title('Senior Citizens',fontsize=14)
plt.legend( bbox_to_anchor=(1, 1))
plt.tight_layout()

for p in ax.patches:
    ax.text(p.get_x()+.15,p.get_height()-100,str(p.get_height()),fontsize=18,color='white',fontweight='bold')
```

568 out of 1142 Senior Citizens are Females.

# 3.Compare 'tenure' with 'Total Charges'

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x=df['tenure'],y=df['TotalCharges'])
plt.title('Tenure vs Total Charges',fontsize=15)
plt.xlabel('Tenure',fontsize=12)
plt.ylabel('Total Charges',fontsize=12)
```

# 4.Find out which contract preferred by the senior citizen.

```python
data2=df[df['SeniorCitizen']==1]

plt.figure(figsize=(8,8))
sns.countplot(data2['Contract'],hue=data2['SeniorCitizen'])
plt.xlabel('Contract',fontsize=12)
plt.ylabel('Count',fontsize=12)
```

Month-to-month is the most preferred type of contract by Senior Citizens.

# 5.Comment on your findings on Payment Method?

```python
plt.figure(figsize=(10,10))
plt.style.use('seaborn')
sns.countplot(df['PaymentMethod'],hue=df['gender'])
plt.xlabel('Payment Method',fontsize=14)
plt.ylabel('Count',fontsize=14)
plt.yticks(fontsize=12)
plt.xticks(fontsize=12)
```

The most preferred payment method is Electronic check. The least preferred payment method by males is Bank transfer and by females is Credit card.
